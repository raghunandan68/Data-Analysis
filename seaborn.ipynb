# Seaborn -Statistical Data Visualization

## What is Seaborn?

Seaborn is a python library that builds on the top of matplotlib that makes the statistical data more easy and beautiful.

## Why we need it?

1. For creating plots like violin plot,box plot and paired plots.

2. Automatically handles data from pandas dataframes.

## Where we use it?

1. Data Analysis and reporting.

2. Exploring relationship between variables.

3. Machine learning for visualizing datasets and results.

## Importing Seaborn

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
sns.histplot(df['age'].dropna(), bins=10, kde=True)
plt.title("Histogram of Age")
plt.show()
 
# Countplot
sns.countplot(x='class', data=df)
plt.title("Count of Passengers by Class")
plt.show()
```

```python
sns.scatterplot(x='age', y='fare', data=df, hue='sex')
plt.title("Age vs Fare")
plt.show()
```

```python
sns.boxplot(x='class', y='age', data=df)
plt.title("Boxplot of Age by Class")
plt.show()
 
# Violin plot
sns.violinplot(x='class', y='age', hue='sex', data=df, split=True)
plt.title("Violin Plot of Age by Class and Sex")
plt.show()
```

```python
iris = sns.load_dataset('iris')
sns.pairplot(iris, hue='species')
plt.show()
```

```python
corr = iris.select_dtypes(include=['float64', 'int64']).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Iris Dataset")
plt.show()
```

```python
sns.set_theme(style="whitegrid")  # options: darkgrid, white, ticks, etc.
sns.histplot(df['age'].dropna(), bins=10, color='orange')
plt.title("Customized Histogram with Seaborn")
plt.show()
```

```python
sns.barplot(x='class', y='fare', data=df, hue='sex')
plt.title("Average Fare by Class and Sex")
plt.show()
```

```python
df=sns.load_dataset("titanic")
sns.barplot(x='class',y='survived',data=df,hue='sex')
plt.title("Passenger survival by Gender and class")
plt.show()
```

```python
sns.histplot(data=df, x='age', bins=30, kde=True, color='skyblue')
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
sns.boxplot(data=df, x='class', y='fare')
plt.title('Fare Distribution by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.show()
```

```python
sns.scatterplot(data=df, x='age', y='fare', hue='class', style='sex')
plt.title('Relationship Between Age and Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
corr = df[['survived', 'pclass', 'age', 'sibsp', 'parch', 'fare']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
