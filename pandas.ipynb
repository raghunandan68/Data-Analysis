# Pandas - Data Analysis Library

## 1. Why Learn Pandas

### What is it?

Pandas is a Python libraray for data manipulation and analysis. 

It provides Series (1D) and DataFrames (2D tabular data).

### Why we need it?

1.Simplifies data cleaning, exploration, and manipulation.

2.Handles large datasets efficiently.

3.Built on top of NumPy for fast computations.

### Where is it used?

1.Data Analysis & Reporting

2.Machine Learning (preprocessing datasets)

3.Finance, Healthcare, Marketing data analysis

4.CSV/Excel/SQL data handling



## 2.Importing Pandas

```python
import pandas as pd
```

## 3.Series & DataFrames

```python
s=pd.Series([10, 20, 30, 40]) 
print("Series:\n", s) 
data = {'Name': ['Alice','Bob','Charlie'], 'Age':[25,30,35]} 
df = pd.DataFrame(data) 
print("\nDataFrame:\n", df)
```

## 4. Reading & Exploring Data

```python
df = pd.read_csv('C:/Users/Raghunandan/Downloads/archive/titanic.csv')
data = {'Name':['Alice','Bob','Charlie'],'Age':[25,30,35],'Sex':['F','M','M']}
df = pd.DataFrame(data)
print(df.head())
print("---")
print(df.tail())
print("---")
print(df.info())
print("---")
print(df.describe())
print("---")
print(df.shape)
```

## 5. Selecting Columns & Rows

```python
# Select column
print(df['Name'])
 
# Select multiple columns
print("---")
print(df[['Name','Age']])
 
# Select row by index
print("---")
print(df.iloc[0])   
print("---")    # first row
print(df.loc[0])     
print("---")   # first row using label
 
# Select subset of rows and columns
print(df.loc[0:1, ['Name','Sex']])
```

## 6. Filtering & Sorting

```python
# Filter rows
print(df[df['Age'] > 28])
# Sort by Age
print(df.sort_values('Age', ascending=False))
```

## 7. Adding & Removing Columns

```python
# Add new column
df['Age_in_5yrs'] = df['Age'] + 5
print(df)
 
# Drop column
df = df.drop('Age_in_5yrs', axis=1)
print(df)
```

## 8. GroupBy & Aggregation

```python
data = {'Name':['Alice','Bob','Charlie','Alice','Bob'],
        'Score':[85,90,95,80,70]}
df = pd.DataFrame(data)
# Group by Name and calculate mean score
grouped = df.groupby('Name').mean()
print(grouped)
```

## 9. Handling Missing Data

```python
data = {'Name':['Alice','Bob','Charlie','David'],
        'Age':[25, None, 35, 40]}
df = pd.DataFrame(data)
# Fill missing value
df['Age'] = df['Age'].fillna(df['Age'].mean())
print(df)
# Drop rows with missing values
# df = df.dropna()
```
