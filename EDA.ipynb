```python
print('Weeks in course column skew : ',df['weeks_in_course'].skew())
print('Weeks in course column kurtosis : ',df['weeks_in_course'].kurtosis())
```

```python
d1=pd.get_dummies(df['prior_programming_experience'],prefix='Experience',drop_first=True)
print(d1)
df=pd.concat([df,d1],axis=1).drop(['prior_programming_experience'],axis=1)
print(df)
```

```python
d2=pd.get_dummies(df['uses_kaggle'],prefix='Use_Kaggle',drop_first=True)
print(d2)
df=pd.concat([df,d2],axis=1).drop(['uses_kaggle'],axis=1)
print(df)
```

```python
d3=pd.get_dummies(df['passed_exam'],prefix='Passed_exam',drop_first=True)
print(d3)
df=pd.concat([df,d3],axis=1).drop(['passed_exam'],axis=1)
print(df)
```

```python
d4=pd.get_dummies(df['participates_in_discussion_forums'],prefix='Participates',drop_first=True)
print(d4)
df=pd.concat([df,d4],axis=1).drop(['participates_in_discussion_forums'],axis=1)
print(df)
```
